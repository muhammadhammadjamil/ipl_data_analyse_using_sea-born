```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\Dell\\Desktop\\matches.csv')
```

```python
df.head(5)
```

```python
matches_per_season = df.groupby('season').id.count()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x = matches_per_season.index, y = matches_per_season)
plt.title("matches per season")
plt.xlabel("season")
plt.ylabel("No of matches")
plt.show()
```

```python
win_matches=pd.crosstab(df['winner'],df['season'])
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('Seasons')
plt.ylabel('Teams')
plt.title('Matches Won Each Season')
sns.heatmap(win_matches ,annot = True, cmap = 'flare')
```
